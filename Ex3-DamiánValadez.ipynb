```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def mat(N, T):
    return np.random.normal(0, 1, (N, T))

example = mat(20, 10) #This is the matrix that we are asked to construct of 20 by 10
print(np.array(example))
```

```python
def eigvals(A):
    return np.linalg.eigvals(A)

def eigs(matrix):
    A = matrix
    At = np.transpose(A)
    
    matrix1 = np.matmul(A, At) #Not dividing by T
    matrix2 = np.matmul(At, A) #Not dividing by N
    
    eigvals1 = eigvals(matrix1)
    eigvals2 = eigvals(matrix2)
    return eigvals1, eigvals2

exampleeigs = eigs(example) #This are the eigenvalues of the example matrices (D * D^t) and (D^t * D)
print(exampleeigs) #The eigenvalues actually do match
```

Given $M$ a $N$ by $T$ random matrix, the eigenvalues of $M_1 = M M^t$ and $M_2 = M^t M$ actually do match. However, while calculating there appear to be some complex values which are numerically zero ($\sim 10^{-16}$) but make it not obvious that the eigenvalues match. We can note this by printing in an ordered (by real part) and side to side way

```python
sortedeigvals1 = np.sort(exampleeigs[0])[::-1]
sortediegvals2 = np.sort(exampleeigs[1])[::-1]
for eigs in zip(sortedeigvals1, sortediegvals2): #Here we print the eigenvalues so we can see that they match
    print(eigs[0], eigs[1])
```

So with the exception of numerical differences, the eigenvalues match.
