```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def mat(N, T):
    return np.random.normal(0, 1, (N, T))
def eigvals(A):
    return np.linalg.eigvals(A)
```

```python
def main(nmem, N, T): #WOE
    xval = np.zeros(nmem * N)
    j = 0
    for i in range(nmem):
        A = mat(N, T)
        At = np.transpose(A)
        matrix = np.matmul(A, At)/T
        eigval = eigvals(matrix)
        
        for k in range(N):
            xval[j] = eigval[k]
            j += 1
    return xval
```

```python
def MPDist(N, T):
    k = T/N
    l = (1/sqrt(k) - 1)**2
    L = (1/sqrt(k) + 1)**2
    X = np.linspace(l, L, 1000)
    Y = []
    for x in X:
        Y.append((k/(2*pi*x)) * sqrt((L - x) * (x - l)))
    return X, Y
```

```python
eigs1015 = main(200, 10, 15)
eigs2025 = main(200, 20, 25)
eigs100200 = main(200, 100, 200)
```

```python
plt.hist(eigs1015, 30, density=True)
plt.plot(MPDist(10, 15)[0], MPDist(10, 15)[1])
```

```python
plt.hist(eigs2025, 30, density=True)
plt.plot(MPDist(20, 25)[0], MPDist(20, 25)[1])
```

```python
plt.hist(eigs100200, 30, density=True)
plt.plot(MPDist(100, 200)[0], MPDist(100, 200)[1])
```
